```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
import pandas as pd
```

```python
a =np.arange(-1, 1, 0.01)
b=np.arange(-1, 1, 0.01)
# b=  np.arange(-10, 10)
a, b = np.meshgrid(a, b)
# print(a)
```

```python
fig= plt.figure()
plot_3d = fig.gca(projection="3d")
plot_3d.plot_surface(a,b,a**2+ b**2-1, cmap='rainbow')
plt.show()
```
